### Ejercicio 3.1 – Calculadora de propina (12 pts)

Escribe un programa que:
1. Pida el **total de la cuenta** (puede ser decimal).
2. Pida el **porcentaje de propina** (número entero).
3. Calcule la **propina** y el **total a pagar**.

**Salida esperada:**
```
Ingresa el total de la cuenta: 8500
Ingresa el porcentaje de propina (%): 10
----------------------------
Propina: $850.0
Total a pagar: $9350.0
```


````python
total = float(input("ingrese el total de la cuenta: "))
propina = int(input("ingrese el porcentaje de propina (%): "))
propina = total / propina
pagar = total - propina
print(pagar)

````

### Ejercicio 3.2 – Clasificador de notas (16 pts)

Escribe un programa que:
1. Pida una **nota entre 1.0 y 7.0**.
2. Valide el rango → si no es válida, muestra `"Nota inválida."` y termina.
3. Muestre el **concepto** y si está **aprobado o reprobado**.

| Rango        | Concepto      |
|--------------|---------------|
| 6.0 – 7.0    | Excelente     |
| 5.0 – 5.9    | Bueno         |
| 4.0 – 4.9    | Suficiente    |
| 1.0 – 3.9    | Insuficiente  |

**Salida esperada:**
```
Ingresa tu nota (1.0 - 7.0): 5.5
Concepto: Bueno
Estado: Aprobado
```


````python
nota = float(input("ingrese la nota: "))
if nota < 1.0 and nota > 7.0:
    print("no es valida")
elif nota >= 6.0:
    print("excelente")
elif nota >= 5.0:
    print("bueno")
elif nota >= 4.0:
    print("suficiente")
else:
    print("insuficiente")
````

### Ejercicio 3.4 – Adivina el número (16 pts)

- El número secreto es `secreto = 42` (fijo, **no lo pidas al usuario**).
- Usa `while` para pedir intentos repetidamente.
- En cada intento indica si el número es **mayor** o **menor**.
- Al adivinar, muestra cuántos intentos necesitó.

**Salida esperada:**
```
Adivina el número secreto (entre 1 y 100):
Intento 1: 50
Es menor. Intenta de nuevo.
Intento 2: 30
Es mayor. Intenta de nuevo.
Intento 3: 42
¡Correcto! Adivinaste en 3 intentos.
```


````python
numero_s = 42
numero_usuario = 0
while numero_s != numero_usuario:
    numero_usuario = int(input("ingrese un numero: "))
    if numero_usuario != numero_s:
        print("error")
print("correcto, el numero secreto era:", numero_s)
````
